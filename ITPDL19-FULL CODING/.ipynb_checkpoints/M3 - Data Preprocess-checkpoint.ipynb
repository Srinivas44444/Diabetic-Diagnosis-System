## DATA PREPROCESSING AND DATA CLEANING

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('customer_activation.CSV')
df = df.drop(columns=['CLIENTNUM'])
df = df.drop(columns=['Income_Category'])
df = df.drop(columns=['Contacts_Count_12_mon'])
df = df.drop(columns=['Avg_Open_To_Buy'])
df = df.drop(columns=['Total_Amt_Chng_Q4_Q1'])
df = df.drop(columns=['Total_Trans_Ct'])
df = df.drop(columns=['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1'])
df = df.drop(columns=['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'])
df = df.drop(columns=['Total_Ct_Chng_Q4_Q1'])
df = df.drop(columns=['Avg_Utilization_Ratio'])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

var = ['Attrition_Flag','Gender','Education_Level','Marital_Status','Card_Category']
  
for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```
